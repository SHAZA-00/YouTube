```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/sample_data/GlobalYouTubeStatistics.csv', encoding='latin1')
df.head()
```

```python
pd.set_option('display.float_format', "{:.2f}".format)
```

```python
category_subscribers_mean = df.groupby('category')['subscribers'].mean()
category_subscribers_mean
```

```python
category_subscribers_mean.sort_values(ascending= False, inplace=True)
```

```python
category_subscribers_mean.head(8).plot(kind= 'bar')
plt.title('Subscribers Mean by Category')
plt.xlabel('Category')
plt.ylabel('Mean Subscribers')
plt.xticks(rotation = 90)
plt.show()
```

```python
channel_type_counts = df['channel_type'].value_counts()
channel_type_counts
```

```python
channel_type_counts.head().plot(kind= 'pie', autopct= '%1.1f%%')
plt.title('Channel Type Counts')
plt.show()
```

```python
plt.scatter(df['subscribers'], df['video views'])
plt.title('Subscribers vs. Video Views')
plt.xlabel('Subscribers')
plt.ylabel('Video Views')
plt.show()
```

```python
country_subscribers_total = df.groupby('Country')['subscribers'].sum().sort_values(ascending= False)
country_subscribers_total
```

```python
country_subscribers_total.head().plot(kind= 'bar')
plt.title('Subscribers Total by Country')
plt.xlabel('Country')
plt.ylabel('Total Subscribers')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.boxplot(x= 'channel_type', y= 'highest_yearly_earnings', data= df)
plt.title('Highes Yearly Earnings by Channel Type')
plt.show()
```

```python
df.groupby(['category', 'Country'])['subscribers'].sum().unstack().plot(kind= 'bar', stacked= True)
plt.title('Subscribers Breakdwon By Category & Country')
plt.xlabel('Category')
plt.ylabel('Subscribers')
plt.legend(title= 'Country', bbox_to_anchor= (1.05, 1), loc= 'upper left')
plt.show()
```

```python
df.groupby(['created_year', 'channel_type'])['subscribers'].mean().unstack().tail().plot(kind= 'bar', width=1)
plt.title('Average Subscribers By created year & channel type')
plt.xlabel('Created Year')
plt.ylabel('Average Subscribers')
plt.legend(title= 'Channel Type', bbox_to_anchor= (1.05, 1), loc= 'upper left')
plt.show()
```

```python
df.groupby(['created_year', 'created_month'])['subscribers'].sum().unstack().plot(kind= 'area', stacked= True)
plt.title('Subscribers Growth Over Time')
plt.xlabel('Year')
plt.ylabel('Subscribers')
plt.legend(title= 'Month', bbox_to_anchor= (1.05, 1), loc= 'upper left')
plt.show()
```

```python
channel_type_subscribers = df.groupby(['channel_type', 'created_year'])['subscribers_for_last_30_days'].sum().unstack()
channel_type_subscribers
```

```python
sns.heatmap(channel_type_subscribers, cmap='YlGnBu', annot= True)
plt.title("Mean Subscribers By created year & channel type")
plt.xlabel('Created Year')
plt.ylabel("Channel Type")
plt.show()
```

```python
df.groupby('created_month')['subscribers_for_last_30_days'].mean().plot(kind= 'line', marker='o')
plt.title('Average Subscribers By created month')
plt.xlabel('Created month')
plt.ylabel('Average Subscribers')
plt.show()
```
